```python
from pathlib import Path
import pandas as pd
file_path = Path("Sleep_health_and_lifestyle_dataset.csv")
```

```python
sleep_health_and_lifestyle_part_1 = pd.read_csv(file_path, encoding="UTF-8")
sleep_health_and_lifestyle_part_1.sort_values("Age")
sleep_health_and_lifestyle_part_1
```

```python
#sleep_health_and_lifestyle.count()
```

```python
sleep_health_and_lifestyle_part_1 = sleep_health_and_lifestyle_part_1.fillna('None')
sleep_health_and_lifestyle_part_1
```

```python
sleep_health_and_lifestyle_part_1.count()
```

```python
file_path = Path("Sleep_health_and_lifestyle_dataset_part_2.csv")
sleep_health_and_lifestyle_part_2 = pd.read_csv(file_path, encoding="UTF-8")
sleep_health_and_lifestyle_part_2.sort_values("Age")
sleep_health_and_lifestyle_part_2
```

```python
sleep_health_and_lifestyle_part_2 = sleep_health_and_lifestyle_part_1.fillna('None')
sleep_health_and_lifestyle_part_2
```

```python
dfs = [sleep_health_and_lifestyle_part_1, sleep_health_and_lifestyle_part_2]
combined_df = pd.concat(dfs)
print(combined_df.count())
```

```python
combined_cleaned_df = combined_df.drop_duplicates()
print(combined_cleaned_df.count())
```

```python
occupations = combined_df['Occupation'].value_counts()
```

```python
occupations.count()
```

```python
occupations.plot.pie()
```

```python
combined_df['Sleep Disorder'].value_counts()
```

```python
sleep_apnea = combined_df[combined_df['Sleep Disorder'] == 'Sleep Apnea']
sleep_apnea.count()
```

```python
insomnia = combined_df[combined_df['Sleep Disorder'] == 'Insomnia']
insomnia.count()
```

```python
occupations_sleep_apnea = sleep_apnea['Occupation'].value_counts()
occupations_sleep_apnea.plot.pie()
```

```python
occupations_insomnia = insomnia['Occupation'].value_counts()
occupations_insomnia.plot.pie()
```
