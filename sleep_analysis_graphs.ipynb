```python
screen_and_sleep = df[["screen time","sleep time"]]

# Set the index to be "State" so they will be used as labels
screen_and_sleep = screen_and_sleep.set_index("screen time")
df.plot(kind="scatter", x="screen time", y="sleep time", grid=True, figsize=(8,8),
              title="Screen Time vs Sleep Time")
plt.show()
```
